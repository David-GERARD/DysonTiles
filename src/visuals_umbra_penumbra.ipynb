```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

sun_position = np.array([0, 0])
sun_diameter = 3.0

earth_position = np.array([20, 0])
earth_diameter = 2

shade_position = np.array([12, 0])
shade_diameter = 0.25

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the sun
plt.plot([sun_position[0], sun_position[0]],
            [sun_position[1]-sun_diameter/2, sun_position[1]+sun_diameter/2], label='Sun', color='orange', linewidth=3)

# Plot the earth
plt.plot([earth_position[0], earth_position[0]],
            [earth_position[1]-earth_diameter/2, earth_position[1]+earth_diameter/2], label='Earth', color='blue', linewidth=3)
# Plot the shade
plt.plot([shade_position[0], shade_position[0]],
            [shade_position[1]-shade_diameter/2, shade_position[1]+shade_diameter/2], label='Shade', color='black', linewidth=3)


plt.ylim(-2, 2)

plt.legend()
```

```python


sun_position = np.array([0, 0])
sun_diameter = 3.0

earth_position = np.array([20, 0])
earth_diameter = 2

shade_position = np.array([12, 0])
shade_diameter = 0.25

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the sun
plt.plot([sun_position[0], sun_position[0]],
            [sun_position[1]-sun_diameter/2, sun_position[1]+sun_diameter/2], label='Sun', color='orange', linewidth=3)

# Plot the earth
plt.plot([earth_position[0], earth_position[0]],
            [earth_position[1]-earth_diameter/2, earth_position[1]+earth_diameter/2], label='Earth', color='blue', linewidth=3)
# Plot the shade
plt.plot([shade_position[0], shade_position[0]],
            [shade_position[1]-shade_diameter/2, shade_position[1]+shade_diameter/2], label='Shade', color='black', linewidth=3)

# Umbra rays
slope_top_ray_umbra = ((shade_position[1] + shade_diameter / 2)-(sun_position[1] + sun_diameter / 2) ) / (shade_position[0] - sun_position[0])
slope_bottom_ray_umbra = ((shade_position[1] - shade_diameter / 2) - (sun_position[1] - sun_diameter / 2)) / (shade_position[0] - sun_position[0])

x_umbra_rays_intercection = -(sun_position[1]+sun_diameter/2-(sun_position[1]-sun_diameter/2)) / (slope_top_ray_umbra - slope_bottom_ray_umbra)

plt.plot([sun_position[0], earth_position[0]],[sun_position[1] + sun_diameter / 2, sun_position[1]+sun_diameter/2 + slope_top_ray_umbra*earth_position[0]], color='orange', linestyle='--')
plt.plot([sun_position[0], earth_position[0]],[sun_position[1] - sun_diameter / 2, sun_position[1]-sun_diameter/2 + slope_bottom_ray_umbra*earth_position[0]], color='orange', linestyle='--')

#Penumbra rays
slope_top_ray_penumbra = ((shade_position[1] + shade_diameter / 2) - (sun_position[1] - sun_diameter / 2)) / (shade_position[0] - sun_position[0])
slope_bottom_ray_penumbra = ((shade_position[1] - shade_diameter / 2) - (sun_position[1] + sun_diameter / 2)) / (shade_position[0] - sun_position[0])
plt.plot([sun_position[0], earth_position[0]],[sun_position[1] - sun_diameter / 2, sun_position[1]-sun_diameter/2 + slope_top_ray_penumbra*earth_position[0]], color='orange', linestyle='--')
plt.plot([sun_position[0], earth_position[0]],[sun_position[1] + sun_diameter / 2, sun_position[1]+sun_diameter/2 + slope_bottom_ray_penumbra*earth_position[0]], color='orange', linestyle='--')

plt.fill_between(x = [shade_position[0],earth_position[0]],
                 y1= [shade_position[1] + shade_diameter / 2, sun_position[1]-sun_diameter/2 + slope_top_ray_penumbra*earth_position[0]],
                 y2= [shade_position[1] - shade_diameter / 2, sun_position[1]+sun_diameter/2 + slope_bottom_ray_penumbra*earth_position[0]],
                 color='gray', alpha=0.5, label='Penumbra')
                 

plt.fill_between(x = [shade_position[0],x_umbra_rays_intercection],
                 y1 = [shade_position[1] + shade_diameter / 2, sun_position[1]+sun_diameter/2 + slope_top_ray_umbra*x_umbra_rays_intercection],
                 y2 = [shade_position[1] - shade_diameter / 2, sun_position[1]-sun_diameter/2 + slope_bottom_ray_umbra*x_umbra_rays_intercection],
                 color='black', alpha=0.8, label='Umbra')
plt.legend()
plt.ylim(-2, 2)
```
